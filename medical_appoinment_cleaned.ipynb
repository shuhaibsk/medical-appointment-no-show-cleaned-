```python
import pandas as pd
import numpy as np

```

```python
df = pd.read_csv('/content/archive (8).zip')
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
df.rename(columns={
    'Hipertension': 'Hypertension',
    'Handcap': 'Handicap',
    'No-show': 'No_show'
}, inplace=True)

df.columns = df.columns.str.lower()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['scheduledday'] = pd.to_datetime(df['scheduledday'])
df['appointmentday'] = pd.to_datetime(df['appointmentday'])
```

```python
df.head()
```

```python
df.info()
```

```python
df['waiting_days'] = (df['appointmentday'] - df['scheduledday']).dt.days
df = df[df['waiting_days'] >= 0]  # remove negative waiting times
```

```python
df.head()
```

```python
df['age'].describe()
df = df[(df['age'] >= 0) & (df['age'] <= 100)]
#for understanding the age gape.
```

```python
#changing no_show column into boolean values
df['no_show'] = df['no_show'].map({'Yes': 1, 'No': 0})
```

```python
df.head()
```

```python
# How many people missed appointments
df['no_show'].value_counts(normalize=True)

```

```python


# Relation between age and no_show
df.groupby('no_show')['age'].mean()



```

```python
# Average waiting days for no-show vs show
df.groupby('no_show')['waiting_days'].mean()
```

```python

df.to_csv('/content/cleaned_medical_appointments.csv', index=False)
```
